```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_excel('GPTW.xlsx')
df.head()
```

```python
headers = df.iloc[2]
df = df[5:]
df.columns = headers
df = df.reset_index(drop=True)
df.head()
```

```python
new_columns = ["Category", "Question", "Overall"]
df.columns = new_columns + list(df.columns[3:])
df.head()
```

```python
df['Category'] = df['Category'].fillna(method='ffill')
df.head()
```

```python
grouped_dfs = {category: group[['Male', 'Female', 'Another gender not listed']].reset_index(drop=True)[:-1]
               for category, group in df.groupby('Category')}
```

```python
qualities = ['Average of All Statements',
    'Camaraderie', 'Caring', 'Collaboration', 'Communication', 
    'Community', 'Competence', 'Corporate Image', 'Credibility', 
    'Engagement', 'Equity', 'Fairness', 'Hospitality', 
    'Impartiality', 'Innovation', 'Integrity', 'Intimacy', 
    'Justice', 'Leadership Behavior', 'Personal Job', 
    'Pride', 'Respect', 'Support', 'Team'
]

filtered_df = df[df['Category'].isin(qualities)]
last_rows = filtered_df.groupby('Category').last().reset_index()
last_rows
```

```python
# Plotting
plt.figure(figsize=(12, 6))
last_rows[['Male', 'Female', 'Another gender not listed']].plot(kind='bar', ax=plt.gca())
plt.title('Last Row Gender-wise Analysis for Each Quality')
plt.xlabel('Category')
plt.ylabel('Scores')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
def plot_pattern_for_series(values: pd.Series, sub_category: str, ylim: tuple):

    plt.figure(figsize=(16, 10))
    plt.plot(values.index, values.values, marker='o')
    plt.title(f'Scores by {sub_category}')
    plt.xlabel(sub_category)
    plt.ylabel('Scores')
    plt.ylim(ylim)  
    plt.grid()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
statements_df = last_rows.iloc[0]
values = statements_df[['Male', 'Female', 'Another gender not listed']]
ylim = (30,80)
plot_pattern_for_series(values, 'Gender', ylim)
```
