```python
import pandas as pd
from textblob import TextBlob

# Load the datasets
values_pulse_data = pd.read_excel('Values Pulse Check 2023_Overall_Raw Data.xlsx')

# Extract relevant open-ended responses
# Assume Q20 is an open-ended response question
text_data = values_pulse_data['Q20'].dropna()  # Drop missing values

# Function to analyze sentiment
def analyze_sentiment(text):
    # Create a TextBlob object
    blob = TextBlob(text)
    # Return the polarity (-1 to 1) and subjectivity (0 to 1)
    return blob.sentiment.polarity, blob.sentiment.subjectivity

# Apply sentiment analysis to each response
sentiment_results = text_data.apply(analyze_sentiment)
sentiment_df = pd.DataFrame(sentiment_results.tolist(), columns=['Polarity', 'Subjectivity'])

# Add sentiment scores to the original data
values_pulse_data = values_pulse_data.loc[text_data.index].reset_index(drop=True)
values_pulse_data['Sentiment_Polarity'] = sentiment_df['Polarity']
values_pulse_data['Sentiment_Subjectivity'] = sentiment_df['Subjectivity']
values_pulse_data

# # Display the updated dataset with sentiment scores
# import ace_tools as tools

# tools.display_dataframe_to_user("Values Pulse Check with Sentiment Analysis", values_pulse_data[['Q20', 'Sentiment_Polarity', 'Sentiment_Subjectivity']])
```

```python
import matplotlib.pyplot as plt

# Plotting sentiment polarity
plt.figure(figsize=(10, 6))
plt.hist(values_pulse_data['Sentiment_Polarity'], bins=30, alpha=0.7, color='blue')
plt.title('Distribution of Sentiment Polarity')
plt.xlabel('Polarity')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```
